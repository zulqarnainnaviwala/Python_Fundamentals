### ATM Service Revamped

```python
cardNumber = "Card Number"
facility = "Facility"
amount = "Amount"
balance = "Balance"
accounts = "Accounts"
pins = "PIN"
currentBalance = "Current Balance"
newBalance = "New Balance"
accountBalances = "Account Balances"

destination = "Dhoraji Bank Limitted"
greeting = f"Welcome to {destination}!!!"
thank = f"Thanks for using {destination}!!!"

facilities = ["B", "D", "W"]
running = True
shutDown = "Shut down"

log = {cardNumber : "", facility : "", amount : "", balance : ""}
data = {accounts : [], pins : [], accountBalances : []}
atmLog = [[cardNumber, facility, amount, balance]]

print(data)
print(atmLog)
```

```python
def ask(param):
    if param == facility:
        return input(f"Press b for balance inquery \nd for cash deposit \nw for withdrawal : ")
    elif param == shutDown:
        return input("Go to Next Transaction??? [Y/N] : ")
    else:
        return input(f"Enter your {param} : ")

def existingAccount(account):
    log[cardNumber] = account.upper()    
    if account.upper() in data[accounts]:
        return True
    else:
        return False

def validatePIN(pin):
    if pin == data[pins][data[accounts].index(log[cardNumber])]:
        return True
    else:
        return False

def getCurrentBalanceByAccount(account):
    log[balance] = data[accountBalances][data[accounts].index(account.upper())]
    return log[balance]

def validateFacility(task):
    if task.upper() in facilities:
        log[facility] = task
        return True
    else:
        log[facility] = "Invalid entry!"
        log[amount] = "Not Entered!"
        log[balance] = "Not Entered!"
        notify(message = "Invalid entry!")
        return False

def notify(key = "", value = 0, message = ""):
    if message != "":
        print(message)
    else:
        print(f"Your {key} is {value}.")

def balanceInquery():
    log[amount] = 0
    notify(currentBalance, log[balance])

def bankDeposit():
    log[amount] = int(ask(amount))
    log[balance] += log[amount]
    notify(newBalance, log[balance])

def validateDrawing(withdrawalAmount):
    if (withdrawalAmount > log[balance]):
        log[amount] = str(withdrawalAmount) + " - Invalid entry!"
        log[balance] = getCurrentBalanceByAccount(log[cardNumber])
        notify(message = "Sorry! Insufficient Balance!!!")
        return False
    else:
        log[amount] = withdrawalAmount
        return True
    
def cashWithdrawal():
    if validateDrawing(int(ask(amount))):
        log[balance] -= log[amount]
        notify(newBalance, log[balance])
    
def facilitate(task):
    if task.upper() == facilities[0]:
        balanceInquery()
    elif task.upper() == facilities[1]:
        bankDeposit()
    elif task.upper() == facilities[2]:
        cashWithdrawal()

def updateBalance():
    data[accountBalances][data[accounts].index(log[cardNumber])] = log[balance]
    
def processInvalidPIN(account):
    log[cardNumber] = f"{account} entered invalid PIN"
    log[facility] = "Not Entered!"
    log[amount] = 0
    log[balance] = getCurrentBalanceByAccount(account)
    notify(message = "Invalid PIN!")
    
def Save(key, value):
    data[key].append(value)
    
def commitTransaction():
    getCurrentBalanceByAccount(log[cardNumber])
    notify(message = greeting)
    if validateFacility(ask(facility)):
        facilitate(log[facility])
        updateBalance()

def createAccount():
    Save(accounts, log[cardNumber])
    Save(pins, ask(pins))
    Save(accountBalances, 0)
    
def maintainLog():
    atmLog.append([log[cardNumber], log[facility], log[amount], log[balance]])

def close():
    if ask(shutDown).upper() != "Y":
        return False
    else:
        return True

def exit():
    maintainLog()
    notify(message=thank)
    return close()

def getStatement(account):
    template = ["Card Number", "Facility", "Amount", "Balance"]
    print(template)
    for transactions in atmLog:
        if (transactions[0] == account.upper()):
            print(transactions)
```

```python
running = True

while running:
    log = {cardNumber : "", facility : "", amount : "", balance : ""}

    if existingAccount(ask(cardNumber)):
        if validatePIN(ask(pins)):
            commitTransaction()
        else:
            processInvalidPIN(log[cardNumber])
    else:
        createAccount()
        commitTransaction()

    running = exit()
```
